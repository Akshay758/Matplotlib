### HISTOGRAM

```python
#USED TO GRAPHICAL REPRESENTATION OF FREQ DISTRIBUTION OF A DATA SET
#THE X AXIS SHOW INTERVALS RANGE OF YOUR DATA
#THE Y AXIS SHOW HOW MANY DATA POINTS FAIL INTO EACH BIN
```

```python
data = [12,13,14,15,16,17,18,12,24,16,18,19,20,23,26,17,18,13,14,19]

plt.figure(figsize=(7,6))
plt.hist(data,bins=6,color="green",edgecolor="black") #BINS = 6 DIVIDE YOUR ENTIRE DATA RANGE INTO 6 EQUAL PARTS
plt.title("history od data")
plt.xlabel("values")
plt.ylabel("frequency")
plt.grid(axis="both", alpha=0.5)     #ALPHA TRANSPARANCY OF GRID LINE
plt.show() 
```

```python
data = [12,13,14,15,16,17,18,12,24,16,18,19,20,23,26,17,18,13,14,19]

plt.figure(figsize=(7,6))
plt.hist(data,bins=6,color="brown",edgecolor="black")
plt.title("histrygram of data")
plt.xlabel("values")
plt.ylabel("frequency")
plt.grid(axis="x", alpha=0.5)
plt.show()
```

```python
#THE MINIMUM VLUE IN YOUR LIST = 10
#THE MaxIMUM VALUE = 26
#RANGE = 21 - 10 = 11 
#BIN WIDTH = 11 DIVIDED BY 6 = 1.83
#MANY VALUES OCCURE REPEATDLY
#HIGHEST FREQ IN MIDDLE RANGE
```

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
data = [12,15,14,10,18,19,12,13,17]
st.probplot(data,dist="norm",plot=plt) 
plt.title("q-q plot of given data") 
plt.show() 
```

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

data = np.random.normal(0,1,100) 
st.probplot(data,dist="norm",plot=plt) 
plt.show()                                                                 
```
